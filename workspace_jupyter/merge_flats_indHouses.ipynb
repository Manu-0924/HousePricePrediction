```python
import pandas as pd
import numpy as np

from sklearn.utils import shuffle
```

```python
flats = pd.read_csv('data_download/flats_cleaned.csv')
independent = pd.read_csv('data_download/house_cleaned.csv')
```

```python
df_combined = pd.concat([flats, independent], ignore_index=True)

# Shuffle the rows
df_shuffled = shuffle(df_combined)
```
